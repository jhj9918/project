```python
#제주도

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os 
import urllib.request
import csv
import pandas as pd 

#크롬 드라이버 위치 
driver = webdriver.Chrome('/Users/yeeun/Downloads/chromedriver 3')

#모든 제주도 숙소 링크 
driver.get('https://www.airbnb.co.kr/s/%EC%A0%9C%EC%A3%BC%EB%8F%84/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&date_picker_type=flexible_dates&flexible_trip_lengths%5B%5D=weekend_trip&source=structured_search_input_header&search_type=filter_change&query=%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%20%EC%A0%9C%EC%A3%BC%ED%8A%B9%EB%B3%84%EC%9E%90%EC%B9%98%EB%8F%84&place_id=ChIJRUDITFTjDDURMb8emNI2vGY&federated_search_session_id=88e9c576-fbf9-4f51-9655-c21d005341ad&pagination_search=true')
time.sleep(10)

img_url = []
name_list=[]
intro_list=[]
#     b_list=[]
review_list=[]
#     p_list=[]
link_list=[]

mList = []
img_index=[]

for page in range(1, 16):

    element=driver.find_element(By.XPATH,'//*[@id="site-content"]')
    element.click()

#이미지 주소 접근                                    
    images = driver.find_elements(By.XPATH,"//*[@id='FMP-target']")
    images = driver.find_elements(By.CSS_SELECTOR,"._6tbg2q")     
    print("===================================")
    print(len(images))
    print("===================================")
    print(page)
    for image in images :
        url = image.get_attribute('src')
        img_url.append(url)
        print(url)
        
#숙소 이름 접근
    names = driver.find_elements(By.CSS_SELECTOR,".t1jojoys.dir.dir-ltr")
    for name in names:
        title = name.text
        name_list.append(title)
        #print(title)
        

#숙소 한줄소개 접근    
    introductions = driver.find_elements(By.CSS_SELECTOR,".t19nnqvo.dir.dir-ltr")
    for introduction in introductions:
        intro = introduction.text
        intro_list.append(intro)
        #print(intro)
        
#침대 갯수 접근        
#     beds = driver.find_elements(By.CSS_SELECTOR,".dir.dir-ltr")
#     for bed in beds:
#         b = bed.text
#         b_list.append(b)
#         print(b)
        
#리뷰 및 평점 접근
    reviews = driver.find_elements(By.CSS_SELECTOR,".t5eq1io.r4a59j5.dir.dir-ltr")
    for review in reviews:
        re = review.get_attribute('aria-label')
        review_list.append(re)
        print(re)
        
#숙소 고유 링크 접근
    links = driver.find_elements(By.CSS_SELECTOR,"a.ln2bl2p.dir.dir-ltr")
    for link in links:
        lnk = link.get_attribute('href')
        link_list.append(lnk)
        #print(lnk)
        
#다음 페이지 접근
    try:
        page_buttons = driver.find_element(By.CSS_SELECTOR,'a._1bfat5l')
        print(page_buttons)

        page_buttons.click() #다음 페이지 (n+1) 에 해당하는 인덱스의 버튼을 클릭합니다
        time.sleep(3)
    except:
        print('#############다음페이지 없음###############')
        break
    
#--------------------

#이미지 다운
img_folder = './final'
 
if not os.path.isdir(img_folder) : # 없으면 새로 생성하는 조건문 
    os.mkdir(img_folder)
    
for index, link in enumerate(img_url) :
    urllib.request.urlretrieve(link, f'./final/jeju_{index}.jpg')
    img_index.append('jeju_'+str(index))

print(img_index)

img_index_jeju= img_index
img_url_jeju=img_url
name_list_jeju=name_list
intro_list_jeju=intro_list
review_list_jeju=review_list
link_list_jeju=link_list

```
